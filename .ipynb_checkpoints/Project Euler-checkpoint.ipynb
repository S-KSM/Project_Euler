```python
#Project Euler 1
import timeit
def div_by_3(x):
    if x % 3 == 0:
        return x
    else:
        return 0

def div_by_5(x):
    if x % 5 == 0:
        if x % 3 != 0:
            return x
        else:
            return 0
    else:
        return 0
 
sum_35 = 0
x = list(range(1,1000))
for number in x:
    sum_35 += div_by_3(number)
    
for number in x:
    sum_35 += div_by_5(number)
print(sum_35)
```

```python
#Project Euler 2
def fib_recursive(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_recursive(n-1) + fib_recursive(n-2)

count = 0
y = 0
fib_list = []
final = 0
while y < 4000001:
    y = fib_recursive(count)
    if y%2 == 0:
        fib_list.append(y)
    
    count += 1
final = sum(fib_list)
print(final)
```

```python
#Project Euler 3
x = 600851475143
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

prime_factors(x)
```

```python
#Project Euler 4
def palindromic(x):
    x = str(x)
    count = 1
    xn=""
    xp = x[:len(x)//2]
    while count < len(x)//2 + 1:
        xn = xn + str(x[-1*count])
        count +=1

    if xp == xn:
        return True
    else:
        return False
    

for i in range(999,100,-1):
    for j in range(999,100,-1):
        y = palindromic(i*j)
        if y == True:
            print(i*j)
            
# faster
b = 999 ** 2
def palinCheck(n):
	return str(n) == str(n)[::-1]
while b > 10000:
	if palinCheck(b) == True:
		a = 999
		while a > 100:
			if b % a == 0 and b / a < 1000:
				print(b)
				raise SystemExit
			a = a - 1
    b = b - 1
```
